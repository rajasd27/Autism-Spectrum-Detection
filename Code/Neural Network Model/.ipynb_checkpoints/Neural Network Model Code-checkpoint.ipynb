```python
import numpy as np
import pandas as pd
from keras import models
from keras import layers
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, mean_squared_error
from matplotlib import pyplot as plt
# Set random seed
np.random.seed(0)
```

```python
df = pd.read_csv('../Datasets/Final_Dataset.csv', index_col=[0])
```

```python
df.head()
```

```python
labels = df.pop('Class/ASD')
y = np.array(labels)
```

```python
X = df.to_numpy()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42, shuffle=True)
```

```python
print(f'We have {X_train.shape[0]} train samples')
print(f'We have {X_test.shape[0]} test samples')
```

```python
def create_network(optimizer='rmsprop', neurons=16):
    network = models.Sequential()
    network.add(layers.Dense(units=neurons, activation='relu', input_shape=(X_train.shape[1],)))
    network.add(layers.Dense(units=neurons, activation='relu'))
    network.add(layers.Dense(units=1, activation='sigmoid'))
    network.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy']) 
    return network
```

```python
neural_network = KerasClassifier(build_fn=create_network, verbose=1)
```

```python
epochs = [10, 20]
batches = [10, 100]
optimizers = ['rmsprop', 'adam']
neurons = [16, 32]

# Create hyperparameter options
hyperparameters = dict(optimizer=optimizers, epochs=epochs, batch_size=batches, neurons=neurons)
```

```python
grid = GridSearchCV(estimator=neural_network, cv=3, param_grid=hyperparameters)
```

```python
grid_result = grid.fit(X_train, y_train)
```

```python
grid_result.best_params_
```

```python
predictions = grid_result.predict(X_test)
```

```python
loss = grid_result.best_estimator_.model.history.history["loss"]
acc = grid_result.best_estimator_.model.history.history["acc"]
```

```python
# Plot the loss function
fig, ax = plt.subplots(1, 1, figsize=(10,6))
ax.plot(np.sqrt(loss), 'r')
plt.xlim((-1, 20))
ax.set_xlabel(r'Epoch', fontsize=20)
ax.set_ylabel(r'Loss', fontsize=20)
ax.tick_params(labelsize=20)

# Plot the accuracy
fig, ax = plt.subplots(1, 1, figsize=(10,6))
ax.plot(np.sqrt(acc), 'r', label='train')
plt.xlim((-1, 20))
ax.set_xlabel(r'Epoch', fontsize=20)
ax.set_ylabel(r'Accuracy', fontsize=20)
ax.tick_params(labelsize=20)
```

```python
print("--------------------- RESULTS OF NEURAL NETWORKS -----------------------")
print("------------------------------------------------------------------------")
print("Accuracy Score: ", accuracy_score(y_test, predictions))
print("------------------------------------------------------------------------")
print("Mean Squared Error: ", mean_squared_error(y_test, predictions))
print("------------------------------------------------------------------------")
print("Confusion Matrix: \n\n",confusion_matrix(y_test, predictions))
print("------------------------------------------------------------------------")
print("Classification Report: \n\n",classification_report(y_test, predictions))
print("------------------------------------------------------------------------")
```
